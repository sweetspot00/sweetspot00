```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\\大四下\\毕业设计\\data\\order.csv", dtype={"areaCode": object, "sn": object})
```

```python
df
```

```python
df_delete_null = df.dropna(axis=0)
df_delete_null
```

```python
#get areacode
area = df_delete_null.drop_duplicates('areaCode')
area = list(area['areaCode'])
#recode
new_code = np.arange(0,len(area),1)
new_code = list(new_code)
areacode_dict = dict(zip(area,new_code))
areacode_dict
```

```python
#get bikecode
sn = df_delete_null.drop_duplicates('sn')
sn = list(sn['sn'])
new_code = np.arange(0,len(sn),1)
new_code = list(new_code)
sncode_dict = dict(zip(sn,new_code))
```

```python
len(sncode_dict)
```

```python
len(sn)
```

```python
#delete bad endtime
df_delete_end = df_delete_null[~df_delete_null['endTime'].isin([0])]
#delete smaller than 2min
df_delete_end['cha'] = df_delete_end['endTime']-df_delete_end['startTime']
df_delete_time = df_delete_end.loc[df_delete_end['cha']>120000]
# df_delete_time
```

```python
#replace area index
for key in areacode_dict.keys():
    df_delete_time['areaCode'].replace(key,areacode_dict[key],inplace = True)
df_delete_time.to_csv("D:/大四下/毕业设计/data/df_delete_time.csv")
df_delete_time
```

```python
#separate May and October
df_5 = df_delete_time[df_delete_time['startTime']<1538323200000]
df_10 = df_delete_time[df_delete_time['startTime']>=1538323200000]
```

```python
df_5
```

```python
#pick morning time
df_5_morning = pd.DataFrame(columns=['areaCode','sn','startTime','endTime','cha'])
for day in range(15):
    m = 1525132800000 + day*86400000
    n = 1525140000000 + day*86400000
    a = df_5[(df_5['startTime']>=m) & (df_5['startTime']<=n)]
    df_5_morning = pd.concat([df_5_morning,a])
df_5_morning.to_csv("D:/大四下/毕业设计/data/df_5_morning.csv")
```

```python
df_10_morning = pd.DataFrame(columns=['areaCode','sn','startTime','endTime','cha'])
for day in range(15):
    m = 1538352000000 + day*86400000
    n = 1538359200000 + day*86400000
    a = df_10[(df_10['startTime']>=m) & (df_10['startTime']<=n)]
    df_10_morning = pd.concat([df_10_morning,a])
df_10_morning.to_csv("D:/大四下/毕业设计/data/df_10_morning.csv")
```

```python
#pick evening time
df_5_evening = pd.DataFrame(columns=['areaCode','sn','startTime','endTime','cha'])
for day in range(15):
    m = 1525165200000 + day*86400000
    n = 1525172400000 + day*86400000
    a = df_5[(df_5['startTime']>=m) & (df_5['startTime']<=n)]
    df_5_evening = pd.concat([df_5_evening,a])
df_10_evening = pd.DataFrame(columns=['areaCode','sn','startTime','endTime','cha'])
for day in range(15):
    m = 1538384400000 + day*86400000
    n = 1538391600000 + day*86400000
    a = df_10[(df_10['startTime']>=m) & (df_10['startTime']<=n)]
    df_10_evening = pd.concat([df_10_evening,a])
df_5_evening.to_csv("D:/大四下/毕业设计/data/df_5_evening.csv")
df_10_evening.to_csv("D:/大四下/毕业设计/data/df_10_evening.csv")
```
