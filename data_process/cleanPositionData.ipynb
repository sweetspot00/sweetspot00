```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("D:\\大四下\\毕业设计\\data\\position.csv",header = None)
```

```python
df_delete_null = df.dropna(axis=0)
# df_delete_null
```

```python
a = (df_delete_null.shape[0])//3
df1 = df_delete_null[0:a]
df2 = df_delete_null[a:2*a]
df3 = df_delete_null[2*a:3*a]
df1
```

```python
df1_s0 = df1[df1[4]==0]
df2_s0 = df2[df2[4]==0]
df3_s0 = df3[df3[4]==0]
status0 = pd.concat([df1_s0,df2_s0,df3_s0])
status0.to_csv("D:/大四下/毕业设计/data/status0.csv")
```

```python
df1_s1 = df1[df1[4]==1]
df2_s1 = df2[df2[4]==1]
df3_s1 = df3[df3[4]==1]
status1 = pd.concat([df1_s1,df2_s1,df3_s1])
# status1.to_csv("D:/大四下/毕业设计/data/status1.csv")
```

```python
# status0.drop(status0[1].str,axis=0,inplace=True)
# status0
status1[1] = status1[1].astype(str)
```

```python
status1.sort_values(1,inplace=True)
status1
```

```python
#find morning time
df_5 = status1[status1[5]<1538323200000]
df_10 = status1[status1[5]>=1538323200000]
df_5_morning = pd.DataFrame([0,1,2,3,4,5])
df_10_morning = pd.DataFrame([0,1,2,3,4,5])
for day in range(15):
    m = 1525132800000 + day*86400000
    n = 1525140000000 + day*86400000
    a = df_5[(df_5[5]>=m) & (df_5[5]<=n)]
    df_5_morning = pd.concat([df_5_morning,a])
df_5_morning = df_5_morning.dropna(axis=0)
df_5_morning.to_csv("D:/大四下/毕业设计/data/log_5_morning.csv")
for day in range(15):
    m = 1538352000000 + day*86400000
    n = 1538359200000 + day*86400000
    a = df_10[(df_10[5]>=m) & (df_10[5]<=n)]
    df_10_morning = pd.concat([df_10_morning,a])
df_10_morning = df_10_morning.dropna(axis=0)
df_10_morning.to_csv("D:/大四下/毕业设计/data/log_10_morning.csv")
```

```python
df_5_morning.sort_values(5,inplace = True)
df_10_morning.sort_values(5,inplace = True)
df_10_morning
```
